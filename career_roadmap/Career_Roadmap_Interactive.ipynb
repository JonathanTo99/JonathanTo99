```python
# ========== ADJUSTABLE PARAMETERS ==========

# Personal Information
BIRTH_YEAR = 1999
GRADUATION_DATE = datetime(2026, 4, 1)  # April 2026

# Phase 1: Industry Lab Work + Prerequisites
PHASE1_DURATION_MONTHS = 21  # 1.5-2 years (adjustable: 18-24 months)

# Phase 2: MS Translational Bioinformatics (University of Utah)
MS_DURATION_YEARS = 2  # (adjustable: 2-3 years)

# Phase 3: PhD Program
PHD_DURATION_YEARS = 5.5  # (adjustable: 5-6 years)

# Phase 4: Postdoc (optional)
SKIP_POSTDOC = True  # Set to False if planning postdoc
POSTDOC_DURATION_YEARS = 2  # Only applies if SKIP_POSTDOC = False

# Financial Parameters
LAB_STARTING_SALARY = 50000  # Research Associate / Lab Tech starting salary
LAB_ANNUAL_RAISE_PCT = 5.0  # Conservative raises for industry lab work
MS_ANNUAL_TUITION = 12000  # University of Utah MS Biomedical Informatics
PHD_STIPEND = 42000
SCIENTIST_STARTING_SALARY = 125000

current_age = datetime.now().year - BIRTH_YEAR
graduation_age = GRADUATION_DATE.year - BIRTH_YEAR

print("üìã Configuration loaded!")
print(f"Current age: {current_age}")
print(f"Age at graduation: {graduation_age}")
```

## üóìÔ∏è Part 2: Calculate Timeline Milestones

```python
def calculate_age(date, birth_year=BIRTH_YEAR):
    """Calculate age at a given date."""
    return date.year - birth_year


def calculate_timeline():
    """Calculate all major career milestones based on configuration parameters."""
    milestones = {}
    
    # Phase 0: Current to Graduation
    milestones['graduation'] = GRADUATION_DATE
    milestones['graduation_age'] = calculate_age(GRADUATION_DATE)
    
    # Phase 1: Industry Lab Work + Prerequisites
    phase1_start = GRADUATION_DATE
    phase1_end = phase1_start + relativedelta(months=PHASE1_DURATION_MONTHS)
    milestones['phase1_start'] = phase1_start
    milestones['phase1_end'] = phase1_end
    milestones['phase1_start_age'] = calculate_age(phase1_start)
    milestones['phase1_end_age'] = calculate_age(phase1_end)
    
    # Phase 2: MS Translational Bioinformatics
    ms_start = phase1_end + relativedelta(months=1)
    ms_end = ms_start + relativedelta(months=int(MS_DURATION_YEARS * 12))
    milestones['ms_start'] = ms_start
    milestones['ms_end'] = ms_end
    milestones['ms_start_age'] = calculate_age(ms_start)
    milestones['ms_end_age'] = calculate_age(ms_end)
    
    # Calculate total lab experience
    lab_years = (ms_end - phase1_start).days / 365.25
    milestones['total_lab_years'] = round(lab_years, 1)
    
    # Phase 3: PhD Applications & Start
    phd_apply_start = ms_end - relativedelta(months=6)
    phd_start = ms_end + relativedelta(months=8)
    phd_end = phd_start + relativedelta(months=int(PHD_DURATION_YEARS * 12))
    milestones['phd_apply_start'] = phd_apply_start
    milestones['phd_start'] = phd_start
    milestones['phd_end'] = phd_end
    milestones['phd_start_age'] = calculate_age(phd_start)
    milestones['phd_end_age'] = calculate_age(phd_end)
    
    # Phase 4: Postdoc (optional)
    if SKIP_POSTDOC:
        career_start = phd_end + relativedelta(months=2)
        milestones['postdoc'] = False
    else:
        postdoc_start = phd_end + relativedelta(months=2)
        postdoc_end = postdoc_start + relativedelta(months=int(POSTDOC_DURATION_YEARS * 12))
        career_start = postdoc_end + relativedelta(months=2)
        milestones['postdoc'] = True
        milestones['postdoc_start'] = postdoc_start
        milestones['postdoc_end'] = postdoc_end
        milestones['postdoc_start_age'] = calculate_age(postdoc_start)
        milestones['postdoc_end_age'] = calculate_age(postdoc_end)
    
    # Phase 5: Industry Career Start
    milestones['industry_start'] = career_start
    milestones['industry_start_age'] = calculate_age(career_start)
    
    # Future projections
    milestones['senior_scientist'] = career_start + relativedelta(years=5)
    milestones['senior_scientist_age'] = calculate_age(milestones['senior_scientist'])
    milestones['principal_scientist'] = career_start + relativedelta(years=10)
    milestones['principal_scientist_age'] = calculate_age(milestones['principal_scientist'])
    
    return milestones


def print_timeline_summary(timeline):
    """Print a formatted summary of key milestones."""
    print("‚úÖ Timeline calculated!")
    print("\nüìÖ Key Milestones:")
    print(f"  ‚Ä¢ Graduation: {timeline['graduation'].strftime('%B %Y')} (age {timeline['graduation_age']})")
    print(f"  ‚Ä¢ MS Completion: {timeline['ms_end'].strftime('%B %Y')} (age {timeline['ms_end_age']})")
    print(f"  ‚Ä¢ Total Lab Experience: {timeline['total_lab_years']} years")
    print(f"  ‚Ä¢ PhD Start: {timeline['phd_start'].strftime('%B %Y')} (age {timeline['phd_start_age']})")
    print(f"  ‚Ä¢ PhD Completion: {timeline['phd_end'].strftime('%B %Y')} (age {timeline['phd_end_age']})")
    print(f"  ‚Ä¢ Industry Career Start: {timeline['industry_start'].strftime('%B %Y')} (age {timeline['industry_start_age']})")


# Calculate and display timeline
timeline = calculate_timeline()
print_timeline_summary(timeline)
```

## üìà Part 3: Interactive Timeline Visualization

```python
def create_phase_data(task, start, finish, resource):
    """Create a phase dictionary for Gantt chart."""
    return dict(Task=task, Start=start, Finish=finish, Resource=resource)


def create_gantt_chart(timeline):
    """Create an interactive Gantt chart showing all career phases."""
    phases = [
        create_phase_data("Phase 0: Finishing BYU", datetime.now(), 
                         timeline['graduation'], "Education"),
        create_phase_data("Phase 1: Lab Work + Prerequisites", 
                         timeline['phase1_start'], timeline['phase1_end'], 
                         "Work + Prerequisites"),
        create_phase_data("Phase 2: MS Translational Bioinformatics (Part-Time)", 
                         timeline['ms_start'], timeline['ms_end'], 
                         "Education + Work"),
        create_phase_data("Phase 3: PhD Applications", 
                         timeline['phd_apply_start'], timeline['phd_start'], 
                         "Transition"),
        create_phase_data("Phase 4: PhD in Oncolytic Virotherapy", 
                         timeline['phd_start'], timeline['phd_end'], 
                         "Education"),
        create_phase_data("Phase 5: Industry Scientist", 
                         timeline['industry_start'], timeline['senior_scientist'], 
                         "Career"),
        create_phase_data("Senior Scientist Level", 
                         timeline['senior_scientist'], timeline['principal_scientist'], 
                         "Career")
    ]
    
    # Add postdoc if applicable
    if timeline['postdoc']:
        phases.insert(-2, create_phase_data(
            "Phase 5: Postdoctoral Training",
            timeline['postdoc_start'], 
            timeline['postdoc_end'], 
            "Research"
        ))
    
    df = pd.DataFrame(phases)
    
    # Define colors
    colors = {
        'Education': '#3498db',
        'Work + Prerequisites': '#e67e22',
        'Education + Work': '#9b59b6',
        'Transition': '#95a5a6',
        'Research': '#1abc9c',
        'Career': '#27ae60'
    }
    
    # Create Gantt chart
    fig = px.timeline(
        df,
        x_start="Start",
        x_end="Finish",
        y="Task",
        color="Resource",
        color_discrete_map=colors,
        title="Career Roadmap Timeline (2025-2045)"
    )
    
    # Set category order
    category_order = [p['Task'] for p in phases]
    
    fig.update_layout(
        height=600,
        xaxis_title="Year",
        yaxis_title="Career Phase",
        showlegend=True,
        hovermode='closest',
        yaxis={'categoryorder': 'array', 'categoryarray': category_order}
    )
    
    return fig


# Create and display Gantt chart
gantt_fig = create_gantt_chart(timeline)
gantt_fig.show()
```

## üí∞ Part 4: Financial Projections

```python
def add_financial_year(year, age, phase, income, tuition, other_expenses):
    """Create a financial data entry for a given year."""
    return {
        'Year': year,
        'Age': age,
        'Phase': phase,
        'Income': income,
        'Tuition': tuition,
        'Other_Expenses': other_expenses,
        'Net': income - tuition - other_expenses
    }


def calculate_finances(timeline):
    """Calculate income, expenses, and net cash flow throughout the timeline."""
    financial_data = []
    
    # Phase 1 & 2: Industry Lab Work
    lab_start_year = timeline['phase1_start'].year
    phd_start_year = timeline['phd_start'].year
    
    current_salary = LAB_STARTING_SALARY
    
    for year in range(lab_start_year, phd_start_year):
        in_ms = timeline['ms_start'].year <= year < timeline['ms_end'].year
        tuition = MS_ANNUAL_TUITION if in_ms else 0
        other_expenses = 4000 if year == lab_start_year else 500
        phase = 'Lab Work + MS' if in_ms else 'Lab Work + Prerequisites'
        
        financial_data.append(add_financial_year(
            year, year - BIRTH_YEAR, phase, 
            current_salary, tuition, other_expenses
        ))
        
        current_salary *= (1 + LAB_ANNUAL_RAISE_PCT / 100)
    
    # Phase 4: PhD Years
    for year in range(phd_start_year, timeline['phd_end'].year + 1):
        financial_data.append(add_financial_year(
            year, year - BIRTH_YEAR, 'PhD Program',
            PHD_STIPEND, 0, 500
        ))
    
    # Phase 5: Postdoc (if applicable)
    if timeline['postdoc']:
        postdoc_start_year = timeline['postdoc_start'].year
        postdoc_end_year = timeline['postdoc_end'].year
        for year in range(postdoc_start_year, postdoc_end_year + 1):
            financial_data.append(add_financial_year(
                year, year - BIRTH_YEAR, 'Postdoc',
                60000, 0, 500
            ))
    
    # Phase 6: Industry Career (first 5 years)
    industry_start_year = timeline['industry_start'].year
    scientist_salary = SCIENTIST_STARTING_SALARY
    
    for year in range(industry_start_year, industry_start_year + 5):
        financial_data.append(add_financial_year(
            year, year - BIRTH_YEAR, 'Industry Scientist',
            scientist_salary, 0, 1000
        ))
        scientist_salary *= 1.05
    
    df_finance = pd.DataFrame(financial_data)
    df_finance['Cumulative_Net'] = df_finance['Net'].cumsum()
    
    return df_finance


# Calculate finances
df_finances = calculate_finances(timeline)

# Display summary
total_education_cost = df_finances['Tuition'].sum() + df_finances['Other_Expenses'].sum()
total_income = df_finances['Income'].sum()
final_net = df_finances['Cumulative_Net'].iloc[-1]

print("üí∞ Financial Summary:")
print(f"  ‚Ä¢ Total Education Investment: ${total_education_cost:,.0f}")
print(f"  ‚Ä¢ Total Income During Training: ${total_income:,.0f}")
print(f"  ‚Ä¢ Cumulative Net by Industry Start: ${final_net:,.0f}")

display(df_finances.head(10))
```

```python
def create_financial_chart(df_finances):
    """Visualize financial trajectory over time."""
    fig = go.Figure()
    
    traces = [
        ('Annual Income', df_finances['Income'], '#27ae60', 'lines+markers', 2),
        ('Net Income (After Expenses)', df_finances['Net'], '#3498db', 'lines+markers', 2),
        ('Cumulative Net Worth', df_finances['Cumulative_Net'], '#e74c3c', 'lines+markers', 3)
    ]
    
    for name, data, color, mode, width in traces:
        fig.add_trace(go.Scatter(
            x=df_finances['Year'],
            y=data,
            name=name,
            mode=mode,
            line=dict(color=color, width=width, dash='dash' if 'Cumulative' in name else None)
        ))
    
    fig.update_layout(
        title='Financial Trajectory Over Career Timeline',
        xaxis_title='Year',
        yaxis_title='Amount ($)',
        height=500,
        hovermode='x unified',
        yaxis=dict(tickformat='$,.0f')
    )
    
    return fig


# Create and display financial chart
fig_finance = create_financial_chart(df_finances)
fig_finance.show()
```

## üéÆ Part 5: Interactive Timeline Adjuster

Use the sliders below to adjust timeline parameters and see how they affect your roadmap in real-time!

```python
# Create interactive widgets
style = {'description_width': '200px'}

phase1_slider = widgets.IntSlider(
    value=PHASE1_DURATION_MONTHS, min=12, max=36, step=3,
    description='Phase 1 Duration (months):',
    style=style, continuous_update=False
)

ms_slider = widgets.FloatSlider(
    value=MS_DURATION_YEARS, min=2.0, max=5.0, step=0.5,
    description='MS Duration (years):',
    style=style, continuous_update=False
)

phd_slider = widgets.FloatSlider(
    value=PHD_DURATION_YEARS, min=4.5, max=7.0, step=0.5,
    description='PhD Duration (years):',
    style=style, continuous_update=False
)

postdoc_checkbox = widgets.Checkbox(
    value=not SKIP_POSTDOC,
    description='Include Postdoc?',
    style=style
)

output = widgets.Output()
validation_output = widgets.Output()


def validate_inputs():
    """Validate timeline parameters and show warnings for unrealistic values."""
    warnings = []
    
    if phase1_slider.value < 15:
        warnings.append("‚ö†Ô∏è Phase 1 < 15 months may not allow enough time for chemistry prerequisites")
    
    if ms_slider.value < 2.0:
        warnings.append("‚ö†Ô∏è MS < 2 years is extremely aggressive for part-time study while working full-time")
    
    if phd_slider.value < 5.0:
        warnings.append("‚ö†Ô∏è PhD < 5 years is ambitious; most programs average 5.5-6 years")
    elif phd_slider.value > 6.5:
        warnings.append("‚ö†Ô∏è PhD > 6.5 years may impact funding and career momentum")
    
    temp_timeline = calculate_timeline()
    lab_years = temp_timeline['total_lab_years']
    
    if lab_years < 2.5:
        warnings.append(f"‚ö†Ô∏è Total lab experience ({lab_years:.1f} years) < 2.5 years may weaken PhD application")
    
    industry_age = temp_timeline['industry_start_age']
    if industry_age > 40:
        warnings.append(f"‚ö†Ô∏è Industry career start at age {industry_age} is quite late; consider shortening timeline")
    
    return warnings


def update_timeline(change):
    """Update timeline when widgets change with validation."""
    global PHASE1_DURATION_MONTHS, MS_DURATION_YEARS, PHD_DURATION_YEARS, SKIP_POSTDOC
    
    PHASE1_DURATION_MONTHS = phase1_slider.value
    MS_DURATION_YEARS = ms_slider.value
    PHD_DURATION_YEARS = phd_slider.value
    SKIP_POSTDOC = not postdoc_checkbox.value
    
    new_timeline = calculate_timeline()
    
    # Display validation warnings
    with validation_output:
        validation_output.clear_output(wait=True)
        warnings = validate_inputs()
        if warnings:
            print("\n".join(warnings))
            print()
    
    # Display timeline update
    with output:
        output.clear_output(wait=True)
        print("üîÑ Updated Timeline:")
        print(f"  ‚Ä¢ MS Completion: {new_timeline['ms_end'].strftime('%B %Y')} (age {new_timeline['ms_end_age']})")
        print(f"  ‚Ä¢ Total Lab Experience: {new_timeline['total_lab_years']} years")
        print(f"  ‚Ä¢ PhD Start: {new_timeline['phd_start'].strftime('%B %Y')} (age {new_timeline['phd_start_age']})")
        print(f"  ‚Ä¢ PhD Completion: {new_timeline['phd_end'].strftime('%B %Y')} (age {new_timeline['phd_end_age']})")
        print(f"  ‚Ä¢ Industry Start: {new_timeline['industry_start'].strftime('%B %Y')} (age {new_timeline['industry_start_age']})")
        print(f"\n  ‚è±Ô∏è  Time to Industry Career: {new_timeline['industry_start'].year - datetime.now().year} years from now")


# Attach update function to widgets
for widget in [phase1_slider, ms_slider, phd_slider, postdoc_checkbox]:
    widget.observe(update_timeline, 'value')

# Display widgets
display(widgets.VBox([
    widgets.HTML("<h3>üéØ Adjust Timeline Parameters</h3>"),
    phase1_slider,
    ms_slider,
    phd_slider,
    postdoc_checkbox,
    validation_output,
    output
]))

# Trigger initial display
update_timeline(None)
```

```python
def format_milestone_row(label, date, age):
    """Format a milestone row for markdown table."""
    return f"| {label} | {date.strftime('%B %Y')} | {age} |"


def export_timeline_summary():
    """Export current timeline configuration as markdown."""
    current_timeline = calculate_timeline()
    
    milestones_table = "\n".join([
        "| Milestone | Date | Age |",
        "|-----------|------|-----|",
        format_milestone_row("Graduation (BS)", current_timeline['graduation'], current_timeline['graduation_age']),
        format_milestone_row("Phase 1 Complete", current_timeline['phase1_end'], current_timeline['phase1_end_age']),
        format_milestone_row("MS Start", current_timeline['ms_start'], current_timeline['ms_start_age']),
        format_milestone_row("MS Completion", current_timeline['ms_end'], current_timeline['ms_end_age']),
        f"| Total Lab Experience | - | {current_timeline['total_lab_years']} years |",
        format_milestone_row("PhD Applications", current_timeline['phd_apply_start'], calculate_age(current_timeline['phd_apply_start'])),
        format_milestone_row("PhD Start", current_timeline['phd_start'], current_timeline['phd_start_age']),
        format_milestone_row("PhD Completion", current_timeline['phd_end'], current_timeline['phd_end_age']),
        format_milestone_row("Industry Career Start", current_timeline['industry_start'], current_timeline['industry_start_age']),
        format_milestone_row("Senior Scientist", current_timeline['senior_scientist'], current_timeline['senior_scientist_age']),
        format_milestone_row("Principal Scientist", current_timeline['principal_scientist'], current_timeline['principal_scientist_age'])
    ])
    
    summary = f"""# Career Roadmap Summary (Updated {datetime.now().strftime('%B %d, %Y')})

## Timeline Configuration
- **Birth Year**: {BIRTH_YEAR}
- **Phase 1 Duration**: {PHASE1_DURATION_MONTHS} months
- **MS Duration**: {MS_DURATION_YEARS} years
- **PhD Duration**: {PHD_DURATION_YEARS} years
- **Postdoc**: {'Yes' if not SKIP_POSTDOC else 'No'}

## Key Milestones

{milestones_table}

## Financial Summary
- **Total Education Investment**: ${total_education_cost:,.0f}
- **Total Income During Training**: ${total_income:,.0f}
- **Net by Industry Start**: ${final_net:,.0f}

---
*Generated from Interactive Career Roadmap Notebook*
"""
    
    return summary


# Create export button
export_button = widgets.Button(
    description='üì• Export Timeline Summary',
    button_style='success',
    tooltip='Export current timeline as markdown'
)

export_output = widgets.Output()


def on_export_click(b):
    """Handle export button click."""
    with export_output:
        export_output.clear_output()
        summary = export_timeline_summary()
        
        filename = f"Career_Roadmap_Summary_{datetime.now().strftime('%Y%m%d')}.md"
        with open(filename, 'w') as f:
            f.write(summary)
        
        print(f"‚úÖ Timeline exported to: {filename}")
        print("\nPreview:")
        display(Markdown(summary[:1000] + "..."))


export_button.on_click(on_export_click)

display(export_button)
display(export_output)
```

```python
# Define scenarios
scenarios = {
    "Current Plan (3.8 years lab)": {
        "phase1": 21, "ms_years": 2, "phd_years": 5.5, "postdoc": False
    },
    "Faster Track (3 years lab)": {
        "phase1": 15, "ms_years": 2, "phd_years": 5, "postdoc": False
    },
    "Extended with Postdoc": {
        "phase1": 21, "ms_years": 3, "phd_years": 6, "postdoc": True
    },
    "Slower MS Track (5.3 years lab)": {
        "phase1": 21, "ms_years": 3, "phd_years": 5.5, "postdoc": False
    }
}


def calculate_scenario(config):
    """Calculate timeline for a specific scenario."""
    global PHASE1_DURATION_MONTHS, MS_DURATION_YEARS, PHD_DURATION_YEARS, SKIP_POSTDOC
    
    PHASE1_DURATION_MONTHS = config['phase1']
    MS_DURATION_YEARS = config['ms_years']
    PHD_DURATION_YEARS = config['phd_years']
    SKIP_POSTDOC = not config['postdoc']
    
    return calculate_timeline()


def create_scenario_comparison():
    """Compare multiple timeline scenarios."""
    comparison_data = []
    
    for scenario_name, config in scenarios.items():
        result = calculate_scenario(config)
        
        comparison_data.append({
            'Scenario': scenario_name,
            'Lab Years': f"{result['total_lab_years']:.1f}",
            'PhD Start Age': result['phd_start_age'],
            'PhD End Age': result['phd_end_age'],
            'Industry Start Age': result['industry_start_age'],
            'Industry Start Year': result['industry_start'].year,
            'Years from Now': result['industry_start'].year - datetime.now().year
        })
    
    return pd.DataFrame(comparison_data)


# Compare scenarios
df_comparison = create_scenario_comparison()

print("üìä Timeline Scenario Comparison:\n")
display(df_comparison)

# Visualize comparison
fig_comparison = px.bar(
    df_comparison,
    x='Scenario',
    y='Industry Start Age',
    color='Years from Now',
    title='Age at Industry Career Start - Scenario Comparison',
    text='Industry Start Age',
    color_continuous_scale='Viridis'
)

fig_comparison.update_traces(textposition='outside')
fig_comparison.update_layout(height=400)
fig_comparison.show()
```
