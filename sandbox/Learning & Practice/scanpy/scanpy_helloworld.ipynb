# Scanpy "Visual First" Hello World

Here is your "Visual First" Hello World. We are going to:

Load a pre-processed dataset (3,000 blood cells).

Look inside the "Container."

Generate the map (UMAP).

```python
import scanpy as sc

# 1. LOAD THE "DIGITAL ORGANISM"
# We are downloading a dataset that is already processed (The "shredded strips" are already taped together).
# This returns an 'AnnData' object. Let's call it 'adata'.
adata = sc.datasets.pbmc3k_processed()

# 2. INSPECT THE ANATOMY
# This prints the structure of the container.
print("--- THE CONTAINER STRUCTURE ---")
print(adata)

# 3. VISUALIZE THE MAP
# We tell the 'camera' (sc.pl) to look at the 'umap' coordinates.
# We color the dots based on the 'louvain' column (which holds the cluster names).
sc.pl.umap(adata, color='louvain', title='PBMC 3K - Cell Types')
```
